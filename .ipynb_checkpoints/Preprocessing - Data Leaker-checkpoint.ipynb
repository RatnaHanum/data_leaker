```python
from scipy import stats
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')

filtered_entries = np.array([True] * len(df))

for col in ['CreditScore',	'Age']:
    zscore = abs(stats.zscore(df[col])) # hitung absolute z-scorenya
    filtered_entries = (zscore < 3) & filtered_entries # keep yang kurang dari 3 absolute z-scorenya
    
df = df[filtered_entries] # filter, cuma ambil yang z-scorenya dibawah 3

print(f'Jumlah baris setelah memfilter outlier: {len(df)}')
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')

filtered_entries = np.array([True] * len(df))
for col in ['CreditScore',	'Age']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    low_limit = Q1 - (IQR * 1.5)
    high_limit = Q3 + (IQR * 1.5)

    filtered_entries = ((df[col] >= low_limit) & (df[col] <= high_limit)) & filtered_entries
    
df = df[filtered_entries]

print(f'Jumlah baris setelah memfilter outlier: {len(df)}')
```

```python
## Setelah hapus outliers
pallet1 = '#00cc66'  # Atau warna spesifik yang Anda inginkan

plt.figure(figsize=(12, 5))
for i in range(0, len(nums)):
    plt.subplot(2, len(nums), i+1)
    sns.boxplot(y=df[nums[i]] ,color=pallet1, orient='h')  
    plt.tight_layout()
plt.show()
```

```python
# distribusi Balance (nilai asli)
sns.kdeplot(df['Balance'])
```

```python
# distribusi gaji (setelah log transformation)
sns.kdeplot(np.log(df['Balance']))
# lebih mendekati distribusi normal!
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# berat dan tinggi kita re-scale
df['Balance_norm'] = MinMaxScaler().fit_transform(df['Balance'].values.reshape(len(df), 1))

# umur dan log gaji kita standarisasi (karena distribusi datanya sudah mendekati normal)
df['Balance_std'] = StandardScaler().fit_transform(df['Balance'].values.reshape(len(df), 1))
```

```python
df[['Balance','Balance_std','Balance_norm']].describe()
```

```python
# label encoder
df['Gender'] = df['Gender'].astype('category').cat.codes
df['HasCrCard'] = df['HasCrCard'].astype('category').cat.codes
df['IsActiveMember'] = df['IsActiveMember'].astype('category').cat.codes
df['Geography'] = df['Geography'].astype('category').cat.codes
```
